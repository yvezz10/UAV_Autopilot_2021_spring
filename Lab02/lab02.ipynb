```python
import cv2
import numpy as np

img = cv2.imread('mj.tif')
h, w = img.shape[:2]
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
print(h,w)
```

```python
a = [0] * 256
for i in range(h):
    for j in range(w):
        k=gray[i,j]
        p = int(k)
        a[p]+=1
print(a)
```

```python
pr=[0]*256
s=0
for i in range(256):
    s = s+ a[i]
    pr[i] = s/563000
print(pr)
```

```python
output=[0]*256
for i in range(256):
    output[i] = int(pr[i]*255)
print(output)
```

```python
new = np.zeros([h,w,1],dtype=np.uint8)
for i in range(h):
    for j in range(w):
        new[i,j] = output[int(gray[i,j])]
```

```python
cv2.imshow('result', new)
cv2.waitKey(0)
cv2.destroyAllWindows()
cv2.waitKey(1)
```
